```python
from numpy import cos, sin, pi, exp, array, arange
import matplotlib.pyplot as plt
```

Входные данные:

```python
x0, z0 = 0, 0
v0 = 1
m = 1
g = 10
alpha = pi/4
assert 0 <= alpha <= pi/2, 'alpha < 0 or alpha > pi/2'
beta = 1e-3
assert beta != 0, 'beta = 0'
```

Функции:

```python
vx0 = v0 * cos(alpha)
vz0 = v0 * sin(alpha)
vx = lambda t: vx0 * exp(-(beta/m)*t)
vz = lambda t: (vz0 + g*m/beta) * exp(-(beta/m)*t) - g*m/beta
# v = lambda t: (vx(t)**2 + vz(t)**2)**0.5
x = lambda t: x0 + m/beta*vx0*(1 - exp(-(beta/m)*t))
z = lambda t: z0 - m/beta*g*t + m/beta*(vz0 + g*m/beta)*(1 - exp(-(beta/m)*t))
```

1ый график

```python
T = arange(x0, x0 + 0.2, 1e-3)
X = array(list(map(x, T)))
Z = array(list(map(z, X)))
plt.plot(T, X)
plt.plot(T, Z, 'r-')
plt.xlabel('Time')
plt.ylabel('Cords')
plt.legend(['x(t)', 'z(x)'])
plt.grid(True)
plt.show()
```

2ой график

```python
VX = array(list(map(vx, T)))
VZ = array(list(map(vz, T)))
plt.plot(T, VX)
plt.plot(T, VZ, 'r-')
plt.xlabel('Time')
plt.ylabel('Speed')
plt.legend(['vx(t)', 'vz(t)'])
plt.grid(True)
plt.show()
```
