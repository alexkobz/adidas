```python
from numpy import array,  append
import matplotlib.pyplot as plt
```

```python
def func(m1, m2):
    x1_0, y1_0 = 0, 1
    x2_0, y2_0 = 0, 0
    vx1_0, vy1_0 = 1, 0
    vx2_0, vy2_0 = 0, 0
    n = 3
    tau = 1/10**n
    m = m1/m2
    X1, Y1 = array([x1_0]), array([y1_0])
    X2, Y2 = array([x2_0]), array([y2_0])
    VX1, VY1 = array([vx1_0]), array([vy1_0])
    VX2, VY2 = array([vx2_0]), array([vy2_0])

    for i in range(10*10**n):
        x1 = X1[-1] + 0.5*tau*VX1[-1]
        y1 = Y1[-1] + 0.5*tau*VY1[-1]
        x2 = X2[-1] + 0.5*tau*VX2[-1]
        y2 = Y2[-1] + 0.5*tau*VY2[-1]
        r = ((X2[-1] - X1[-1])**2 + (Y2[-1] - Y1[-1])**2)**0.5
        vx1 = VX1[-1] + 0.5*tau*(X2[-1] + X1[-1])/r**3
        vy1 = VY1[-1] + 0.5*tau*(Y2[-1] + Y1[-1])/r**3
        vx2 = VX2[-1] - 0.5*tau*m*(X2[-1] - X1[-1])/r**3
        vy2 = VY2[-1] - 0.5*tau*m*(Y2[-1] - Y1[-1])/r**3

        X1 = append(X1, [X1[-1] + tau*vx1])
        Y1 = append(Y1, [Y1[-1] + tau*vy1])
        X2 = append(X2, [X2[-1] + tau*vx2])
        Y2 = append(Y2, [Y2[-1] + tau*vy2])
        r = ((x2 - x1)**2 + (y2 - y1)**2)**0.5
        VX1 = append(VX1, [VX1[-1] + tau*(x2 - x1)/r**3])
        VY1 = append(VY1, [VY1[-1] + tau*(y2 - y1)/r**3])
        VX2 = append(VX2, [VX2[-1] - tau*m*(x2 - x1)/r**3])
        VY2 = append(VY2, [VY2[-1] - tau*m*(y2 - y1)/r**3])
    plt.plot(X1, Y1)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('Движение тела на фазовом пространстве')
    plt.grid(True)
    plt.show()
    plt.plot(X2, Y2)
    plt.xlabel('Время')
    plt.ylabel('Скорость')
    plt.title('Движение тела на фазовом пространстве')
    plt.grid(True)
    plt.show()
```

```python
m1, m2 = 1, 1000
func(m1, m2)
```
