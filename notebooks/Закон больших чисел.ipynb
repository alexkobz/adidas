# Закон больших чисел

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.randint(2, size=1000)
Y = np.array([])
tmp = []
for i, j in enumerate(X):
    tmp.append(j)
    Y = np.append(Y, [sum(tmp)/(i+1)*100])
avg = np.count_nonzero(X == 0)/len(X)*100
```

```python
plt.suptitle('Закон больших чисел')
ax = plt.gca()
plt.plot(Y)
plt.yticks(np.arange(0, 101, 10))
plt.axhline(y=50, color='r', linestyle='-')
plt.axhline(y=avg, color='g', linestyle='--')
plt.grid(True)
plt.show()
```
