# Центральная предельная теорема

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
n = 30
```

Для равномерного распределения

```python
uniform = np.array([np.array([np.random.uniform() for _ in range(n)]) for i in range(N)])
```

```python
avg_u = np.array([np.average(i) for i in uniform])
avg_allu = np.average(avg_u)
std_allu = np.std(avg_u)
std_err = std_allu/np.sqrt(N)
uni_norm = [(i-avg_allu)/std_allu for i in avg_u]
```

```python
plt.hist(avg_u)
plt.show()
plt.hist(uni_norm)
plt.show()
```

Для распределения Пуассона

```python
poisson = np.array([np.array([np.random.poisson() for _ in range(n)]) for i in range(N)])
```

```python
avg_p = np.array([np.average(i) for i in poisson])
avg_allp = np.average(avg_p)
std_allp = np.std(avg_p)
std_err = std_allp/np.sqrt(N)
pois_norm = [(i-avg_allp)/std_allp for i in avg_p]
```

```python
plt.hist(avg_p)
plt.show()
plt.hist(pois_norm)
plt.show()
```
