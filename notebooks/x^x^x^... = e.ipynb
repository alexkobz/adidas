# x^x^x^... = e

```python
from math import exp, e
from numpy import array, append
import matplotlib.pyplot as plt
```

```python
x = exp(exp(-1))
X = array(x)
s = f'{x}**'
for _ in range(100):
    s += f'{x}'
    a = eval(s)
    s += '**'
    X = append(X, [a])
```

```python
fig, ax = plt.subplots(1, 1)
plt.plot(X)
plt.title('x^x^x... = e')
plt.axhline(y=e, color='r', linestyle='-')
ax.set_yticks([0, 1, x, 2, e, 3])
ax.set_yticklabels(['0', '1', 'x', '2', 'e', '3'])
plt.grid(True)
plt.show()
```
